```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
from scipy.interpolate import CubicSpline  
from mpl_toolkits.mplot3d import Axes3D

def heart_curve(t):
    x = 3**(0.5)*np.cos(t)
    y = 2*np.sqrt(3**(0.5)*abs(np.cos(t)))/3+2*3**(0.5)*np.sin(t)/3
    return x, y

t = np.linspace(0,2*np.pi,1000)
x,y = heart_curve(t)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Heart Curve',color='red')

t = np.linspace(0, 2*np.pi, 30)
x, y = heart_curve(t)
plt.plot(x, y, label='Isometric fitting', color='blue')
plt.legend()
plt.show()

def curve_fitting_periodic(x,y):
    s = []
    cnt = 0
    for i in range(30):
        if i == 0:
            s.append(0)
        else:
            cnt = cnt + ((x[i]-x[i-1])**2 + (y[i]-y[i-1])**2)**(0.5)
            s.append(cnt)
    cs1 = CubicSpline(s, x, bc_type=('periodic'))
    cs2 = CubicSpline(s, y, bc_type=('periodic'))
    s_new = np.linspace(0, s[29], 1000)
    x_new = cs1(s_new)
    y_new = cs2(s_new)
    plt.figure(figsize=(10, 6))
    plt.plot(x_new, y_new, label='Curve fitting', color='blue')

curve_fitting_periodic(x,y)
t = np.linspace(0,2*np.pi,1000)
x,y = heart_curve(t)
plt.plot(x, y, label='Heart Curve',color='red')
plt.legend()
plt.show()
```

```python
def curve_2(t):
    x = np.sin(t)+t*np.cos(t)
    y = np.cos(t)-t*np.sin(t)
    return x,y
t = np.linspace(0, 6*np.pi, 1000)
x,y = curve_2(t)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Curve', color='red')

t = np.linspace(0, 6*np.pi, 60)
x,y = curve_2(t)
plt.plot(x, y, label='Isometric fitting', color='blue')
plt.legend()
plt.show()

def curve_fitting(x,y):
    s = []
    cnt = 0
    for i in range(60):
        if i == 0:
            s.append(0)
        else:
            cnt = cnt + ((x[i]-x[i-1])**2 + (y[i]-y[i-1])**2)**(0.5)
            s.append(cnt)
    cs1 = CubicSpline(s, x, bc_type=('clamped'))
    cs2 = CubicSpline(s, y, bc_type=('clamped'))
    s_new = np.linspace(0, s[59], 1000)
    x_new = cs1(s_new)
    y_new = cs2(s_new)
    plt.figure(figsize=(10, 6))
    plt.plot(x_new, y_new, label='Curve fitting', color='blue')

curve_fitting(x,y)
t = np.linspace(0, 6*np.pi, 1000)
x,y = curve_2(t)
plt.plot(x, y, label='Curve', color='red')
plt.legend()
plt.show()
```

```python
def curve_3(t):
    x = np.sin(np.cos(t))*np.cos(np.sin(t))
    y = np.sin(np.cos(t))*np.sin(np.sin(t))
    z = np.cos(np.cos(t))
    return x,y,z

t = np.linspace(0,2*np.pi,1000)
x,y,z = curve_3(t)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z,label='curve',color='red')

t = np.linspace(0,2*np.pi,30)
x,y,z = curve_3(t)
ax.plot(x, y, z,label='Isometric fitting',color='blue')
ax.legend()
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

def curve_fitting(x,y,z):
    s = []
    cnt = 0
    for i in range(30):
        if i == 0:
            s.append(0)
        else:
            cnt = cnt + ((x[i]-x[i-1])**2 + (y[i]-y[i-1])**2 + (z[i]-z[i-1])**2)**(0.5)
            s.append(cnt)
    cs1 = CubicSpline(s, x, bc_type=('clamped'))
    cs2 = CubicSpline(s, y, bc_type=('clamped'))
    cs3 = CubicSpline(s, z, bc_type=('clamped'))
    s_new = np.linspace(0, s[29], 1000)
    x_new = cs1(s_new)
    y_new = cs2(s_new)
    z_new = cs3(s_new)
    ax.plot(x_new,y_new,z_new,label='curve fitting',color='blue')
    
curve_fitting(x,y,z)

t = np.linspace(0,2*np.pi,1000)
x,y,z = curve_3(t)
ax.plot(x, y, z,label='curve',color='red')
ax.legend()
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
plt.show()
```
