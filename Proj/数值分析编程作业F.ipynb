```python
import matplotlib.pyplot as plt
import numpy as np

def truncated_power_function(t,x):
    if x<=t:
        return t-x
    else:
        return 0

# 
fig, axs = plt.subplots(3, 3, figsize=(15, 8))

# 
def plot_subfigure(ax, func, x):
    y = np.zeros_like(x)
    n = x.shape[0]
    for i in range(n):
        y[i] = func(x[i])
    ax.plot(x, y)
    ax.set_xlim(x[0], x[n-1])
    ax.set_ylim(0, 1.1)

x = np.linspace(-2,1,300)
def f1(x):
    return truncated_power_function(-1,x)
plot_subfigure(axs[0, 0], f1, x)

def f2(x):
    return truncated_power_function(0,x)
plot_subfigure(axs[1, 0], f2, x)

def f3(x):
    return truncated_power_function(1,x)
plot_subfigure(axs[2, 0], f3, x)

def f4(x):
    return f2(x)-f1(x)
plot_subfigure(axs[1, 1], f4, x)

def f5(x): 
    return f3(x)-f2(x)
plot_subfigure(axs[2, 1], f5, x)

def f6(x):
    return f5(x)-f4(x)
plot_subfigure(axs[2, 2], f6, x)

# 
plt.tight_layout()
plt.show()
```

```python
# 
fig, axs = plt.subplots(4, 4, figsize=(15, 8))

def truncated_power_function(t,x):
    if x<=t:
        return (t-x)**2
    else:
        return 0

# 
def plot_subfigure(ax, func, x):
    y = np.zeros_like(x)
    n = x.shape[0]
    for i in range(n):
        y[i] = func(x[i])
    ax.plot(x, y)
    ax.set_xlim(x[0], x[n-1])
    ax.set_ylim(0, 1.1)

x = np.linspace(-3,1,400)
def f1(x):
    return truncated_power_function(-2,x)
plot_subfigure(axs[0, 0], f1, x)

def f2(x):
    return truncated_power_function(-1,x)
plot_subfigure(axs[1, 0], f2, x)

def f3(x):
    return truncated_power_function(0,x)
plot_subfigure(axs[2, 0], f3, x)

def f4(x):
    return truncated_power_function(1,x)
plot_subfigure(axs[3, 0], f4, x)

def f5(x): 
    return f2(x)-f1(x)
plot_subfigure(axs[1, 1], f5, x)

def f6(x):
    return f3(x)-f2(x)
plot_subfigure(axs[2, 1], f6, x)

def f7(x):
    return f4(x)-f3(x)
plot_subfigure(axs[3, 1], f7, x)

def f8(x):
    return (f6(x)-f5(x))/2
plot_subfigure(axs[2, 2], f8, x)

def f9(x):
    return (f7(x)-f6(x))/2
plot_subfigure(axs[3, 2], f9, x)

def f10(x):
    return f9(x)-f8(x)
plot_subfigure(axs[3, 3], f10, x)

# 
plt.tight_layout()
plt.show()
```
